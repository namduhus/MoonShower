## 문자 안내 출력

```python
cities.values # 도시명 전체 조회
```

```python
# 3단계: 도시명을 입력하면 위도와 별자리리스트와 메시지를 반환하는 함수
 
def predict_best_meteor_shower_viewing(city):
    meteor_shower_string=''
 
    # 입력한 도시가 없으면 오류 메시지 //
    if city not in cities.values:
        meteor_shower_string = "아쉽지만 " + city + "에서는 현재 유성우 관측이 어렵습니다."
        return meteor_shower_string
 
    # 도시의 위도 정보 받아오기
    latitude = cities.loc[cities['city'] == city, 'latitude'].iloc[0]
 
    # 해당도시에서 관측 가능한 별자리 리스트 구하기
    constellation_list = constellations.loc[(constellations['latitudestart']>=latitude) & (constellations['latitudeend']<=latitude), 'constellation'].tolist()
 
    # 별자리가 존재하지 않는다면 오류메시지 //
    if not constellation_list:
        meteor_shower_string = "아쉽지만 " + city + "에서는 볼 수 있는 별자리가 없습니다."
        return meteor_shower_string
 
    # 관측 가능한 별자리를 안내 //
    # 'Wellington에서는 별자리 관측이 가능합니다.'
    meteor_shower_string = city + "에서는 별자리 관측이 가능합니다."
   
    return latitude, constellation_list, meteor_shower_string
```

## 최적의 유성우 관측일

```python
# 도시명을 입력하면 관측가능한 유성우에 대한 정보를 알려주는 함수 작성하기

def predict_best_meteor_shower_viewing(city):
    # 안내 메시지 초기화
    meteor_shower_string = ''

    # 입력한 도시 정보가 존재하지 않으면 오류 메시지
    if city not in cities['city'].values:
        meteor_shower_string = f'{city} 도시 정보를 찾을 수 없습니다.'
        return meteor_shower_string

    # 도시의 위도 정보를 불러오기
    latitude = cities.loc[cities['city'] == city, 'latitude'].iloc[0]

    # 해당 도시의 위도에서 관측 가능한 별자리 조회
    constellations_list = constellations.loc[(constellations['latitudestart'] >= latitude) & (constellations['latitudeend'] <= latitude), 'constellation'].to_list()

    # 볼 수 있는 별자리가 존재하지 않으면 오류 메시지
    if not constellations_list:
        meteor_shower_string = f'{city}에서 관측 가능한 별자리가 없습니다.'
        return meteor_shower_string

    # 별자리 관측 가능 안내 메시지
    meteor_shower_string = f'{city}에서는 별자리 관측이 가능합니다.\n\n'

    # 도시별 관찰 가능한 별자리 iterate
    for constellation in constellations_list:
        # 별자리와 가장 가까운 유성우 조회
        name = meteor_showers.loc[meteor_showers['radiant'] == constellation, 'name'].iloc[0]

        # 유성우 관측이 가능한 시작일과 종료일
        startdate = meteor_showers.loc[meteor_showers['radiant'] == constellation, 'startdate'].iloc[0]
        enddate = meteor_showers.loc[meteor_showers['radiant'] == constellation, 'enddate'].iloc[0]

        # 유성우가 보일 때의 달의 위상 조회
        moon_phase_list = moon_phases.loc[(moon_phases['date'] >= startdate) & (moon_phases['date'] <= enddate)]

        # 유성우가 가장 잘 보이는 날의 첫째날 조회
        best_moon_date = moon_phase_list.loc[moon_phase_list['percentage'].idxmin()]['date']

        # 사용자에게 정보전달
        meteor_shower_string += f'{name}를 잘 보려면 {constellation}자리 위치를 향하여 {best_moon_date.strftime('%Y-%m-%d')}에 보시면 됩니다.\n'

    return meteor_shower_string
```
